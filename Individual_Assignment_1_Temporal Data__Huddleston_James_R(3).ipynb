```python
#  Visualization: 2020SU_MSDS_455-DL_SEC55
#  Individual_Assignment_1_Temporal Data
#  Huddleston_James
#  July 7, 2020 
import os
import pandas as pd
import matplotlib.pyplot as plt
import io
import requests
```

```python
#os.chdir('C:/Users/huddl/OneDrive/A_Northwestern University/MSDS_455_Visualization/Mod3/Assignment')
#df1=pd.read_csv('oilProd1.csv')
# read dataset from github
df=pd.read_csv('https://raw.githubusercontent.com/jrhudlstn/DataSets/master/oilProd1.csv')
```

```python
columns = [ 'OH', 'IL-IN', 'CA', 'KS-NE', 'CO-NM-AZUT', 'TX', 'LA', 'OK']
```

```python
# colors for the columns
colors = ["darkmagenta", "royalblue", 'c', 'k', 'gold', 'r','g', 'burlywood' ]
```

```python
# set frame
plt.figure(figsize=(12,8))
polys = plt.stackplot(df['Year'], df[columns].values.T, colors=colors)
#
rectangles =[]
for poly in polys:
    rectangles.append(plt.Rectangle((0,0),1,1, fc=poly.get_facecolor()[0]))
    legend=plt.legend(rectangles, columns, loc=0)
    frame=legend.get_frame()
    frame.set_color('white')
# set the graph cartesian plane (X,Y proportions and title of the Graph)
plt.title('United States of America Oil Production by States for Years\n1862-2010', fontsize=16)
#plt.xlabel('Year', fontsize=14)
plt.xlabel('Year\n\nCombined annual crude oil production (in US Dollars per year) from the states of Ohio, Illinois, Indiana, California, Kansas, Nebraska,\n Colorado, New Mexico, Arizona, Utah, Texas, Louisiana, and Oklahoma.\n\nSource: Oil Prices, Exhaustible Resources, and Economic Growth. January 2012. James D. Hamilton. University of California, San Diego.\nhttps://www.researchgate.net/publication/228268850_Oil_Prices_Exhaustible_Resources_and_Economic_Growth', fontsize=12)
plt.ylabel('Annual crude oil production (in US Dollars per year)', fontsize=14)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlim(1896, 2010)    
# show figure 1
plt.show()
```
